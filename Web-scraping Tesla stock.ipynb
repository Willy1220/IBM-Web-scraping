```python
# Using request.get() to extract data form web
url="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data = requests.get(url).text
# Parsing data by BeautifulSoup and asign it into soup object
soup = BeautifulSoup(data) 

# Analyzing HTML and breaking it down to find the contents I want which are in "tr" tags
tbody = soup.find_all('tbody')
# The second part of tbody is quarterly revenue
secoend_tbody = tbody[1]
tr2 = secoend_tbody.find_all('tr') # Quarterly revenue

lst_revenue=[]
for row in tr2:
    tds = row.find_all('td')
    date = tds[0].text
    price = tds[1].text
    lst_revenue.append({"Date":date, "Revenue":price})
revenue_data_tesla = pd.DataFrame(lst_revenue)

# Data cleaning (coma and dollar sign)
revenue_data_tesla['Revenue'] = revenue_data_tesla['Revenue'].str.replace(',|\$',"", regex=True) 
# Remove null or empty strings
revenue_data_tesla.dropna(inplace=True)
# using (quarter_data['Revenue'] != "") to extract the values that are not empty
revenue_data_tesla= revenue_data_tesla[revenue_data_tesla['Revenue'] != ""]

revenue_data_tesla
```
